```python
new_dataset = pd.read_csv('Fifa_world_cup_matches.csv')
new_dataset
```

```python
new_dataset.info()
```

```python
matches = ['date','hour','team1','number of goals team1','number of goals team2','team2','category']
matches_df= new_dataset.loc[:, matches]
```

```python
matches_df
```

```python
grupos = new_dataset.groupby('category')
grupo_a = grupos.get_group('Group A')
```

```python
grupo_a_filtrado= grupo_a[matches]
```

```python
grupo_a_filtrado
```

```python
for grupo, datos in grupos:
    matches_por_group = datos[matches]
    print("Grupo:", grupo)
    print(matches_por_group.to_string(index=False))
    print("\n")
```

```python
for grupo, datos in grupos:
    if "Group" in grupo:
        matches_por_group = datos[matches]
        print("Grupo:", grupo)
        print(matches_por_group.to_string(index=False))
        print("\n")
```
